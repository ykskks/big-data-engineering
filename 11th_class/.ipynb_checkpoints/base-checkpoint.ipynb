```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.covariance import EllipticEnvelope
from sklearn.neighbors import LocalOutlierFactor
from sklearn.svm import OneClassSVM
```

```python
def robust_covariance(X):
    rob_cov = EllipticEnvelope(random_state=42)
    return rob_cov.fit_predict(X)
```

```python
def lof(X):
    lof = LocalOutlierFactor(n_neighbors=20)
    return lof.fit_predict(X)
```

```python
def svm(X):
    svm = OneClassSVM(gamma=0.1, nu=0.1, random_state=42)
    return svm.fit_predict(X)
```

```python
data = [pd.read_csv('data/' + fname, header=None).values for fname in ['1.csv', '2.csv']]
methods = [
    ('Robust covariance', robust_covariance),
    ('LOF', lof),
    ('One class SVM', svm)
]
fig, axes = plt.subplots(len(data), len(methods), figsize=(20, 10))

for i, X in enumerate(data):
    for j, (desc, met) in enumerate(methods):
        ax = axes[i, j]
        if i == 0:
            ax.set_title(desc, size=18)
        pred = met(X)
        ax.scatter(x=X[:, 0], y=X[:, 1], c=pred, cmap='Set1')
```
